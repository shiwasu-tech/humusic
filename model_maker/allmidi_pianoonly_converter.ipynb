```python
import os
from mido import MidiFile
from pathlib import Path

# 指定されたディレクトリ内のすべてのMIDIファイルを処理
directory = Path('../datasets/orchestra/')  # ここにディレクトリのパスを指定
save_directory = '../datasets/piano1_only'  # ここに保存先のディレクトリのパスを指定


for filepath in directory.glob("**/*.mid"):
    print(filepath)
    mid = MidiFile(filepath)

    # piano1トラックを抽出
    piano1_track = None
    for i, track in enumerate(mid.tracks):
        for msg in track:
            if msg.type == 'program_change' and msg.program == 0:  # piano1のプログラム番号が0と仮定
                piano1_track = track
                break
        if piano1_track:
            break

    # piano1トラックが見つかった場合、新しいMIDIファイルとして保存
    if piano1_track:
        new_mid = MidiFile()
        new_mid.tracks.append(piano1_track)
        new_filename = os.path.join(save_directory, f'piano1_only_{filepath.name}')
        new_mid.save(new_filename)
        print(f'Saved: {new_filename}')
    else:
        print(f'Piano1 track not found in {filepath.name}')
```
